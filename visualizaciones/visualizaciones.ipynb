```
import pandas as pd
import mplfinance as mpf
```

```
df = pd.read_csv("SAN.csv")
```

```
df
```

```
df.info()
```

```
df.columns
```

```
df.Date = pd.to_datetime(df.Date)
df.info()
```

```
df = df.set_index('Date')
```

```
df
```

```
mpf.plot(df)
```

```
mpf.plot(df, type='line', volume=True)
```

```
mpf.plot(df['2020'], volume=True)
```

```
mpf.plot(df['2020-03'], volume=True)
```

```
mpf.plot(df['2020-01'], type='candle', volume=True)
```

```
mpf.plot(df['2020-01':'2020-03'], type='candle', volume=True)
```

```
mpf.plot(df['2020-01':'2020-03'], type='candle', mav=(10), volume=True)
```

```
mpf.plot(df['2020-01':'2020-03'], type='candle', volume=True, mav=(5), tight_layout=True)
```

```
mpf.plot(
    df['2020-01':'2020-03'], figratio=(20,12), 
    type='candle', 
    title='Precio del Santander', 
    volume=True, mav=(5), 
    tight_layout=True)
```

```
mpf.plot(
    df['2020-01':'2020-03'], figratio=(20,12), 
    type='candle', 
    title='Precio del Santander', 
    volume=True, mav=(5), 
    tight_layout=True,
    style='yahoo')
```
